```python
#dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timezone
pd.set_option('display.max_columns', 50)
```

```python
#import csv data files
# File to Load 
ad_data_to_load = "csv_data/Ad3.csv"
cost_of_goods_to_load = "csv_data/cost_of_goods.csv"
transactions_to_load="csv_data/transactions_main.csv"

# Read the City and Ride Data
ad_data_raw =pd.read_csv(ad_data_to_load, index_col=False)
cost_of_goods_raw =pd.read_csv(cost_of_goods_to_load)
transactions_raw =pd.read_csv(transactions_to_load)

#create dfs
ad_df_raw = pd.DataFrame(ad_data_raw)
cog_df_raw = pd.DataFrame(cost_of_goods_raw)
trans_df_raw = pd.DataFrame(transactions_raw)

#merge trans df and cost of goods df
combo_df =pd.merge(trans_df_raw, cog_df_raw, how='inner', on='sku', left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
#cog_df_raw.head()
combo_df.head(3)
```

```python
combo_df.loc[:, 'date/time'].replace(regex=True,inplace=True,to_replace="PST", value="") #remove 'PST' from date/time; if PST or PDT remain an error will occur
combo_df.loc[:, 'date/time'].replace(regex=True,inplace=True,to_replace="PDT", value="") #remove 'PDT' from date/time; if PST or PDT remain an error will occur
combo_df['date/time']=pd.to_datetime(combo_df['date/time']) #keeps changes in place

combo_df.sort_values('date/time').head()
```

```python
def trans_cols(x): #creates new cols and their calculations
    x['transaction revenue']= ""#revenue generate from transaction (revenue - cost of goods)
    x['sale price']= x['product sales']/x['quantity']#generates price of item at sale
    x['Day of Week']=""
    x['Cost of Goods']=""
    x['Whole Date'] =""
    
trans_cols(combo_df) #apply the trans_cols function

short_trans_df=combo_df.drop(['settlement id','order id','marketplace','fulfillment'],axis=1) #drop unimportant columns and create a smaller df
short_trans_df.head()
```
